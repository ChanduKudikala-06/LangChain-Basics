```python
import os
from dotenv import load_dotenv

load_dotenv

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")

#Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_v2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
#ChatOpenAI is a class from langchain OpenAI integration that allows you to call OPENAI chat models esaily
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
#Input and get the response from LLM

result=llm.invoke("What is Genrative AI")


print(result)
```

```python
#Chat Prompt Template is a structured way to create prompts for chat based llms liuke chatgpt,openai gpt models
#where prompt is dived into different message roles such as system,user,assistant.
#Instead of passing manually it will pass as dynamically.


from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are an AI expert.Provide me answers based on the questions"),
        ("user","{input}")
    ]
)

prompt
```

```python

#Whenever user will give input it will go to prompt and then our llm model
chain=prompt|llm

response=chain.invoke({"input":"What is Langsmith"})

response
```

```python
type(response)
```

```python
#StrOutput Parser it extracts only the text content from the model output
from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"input":"What is Langsmith"})

print(response)
```
