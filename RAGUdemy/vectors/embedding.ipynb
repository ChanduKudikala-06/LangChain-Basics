```python
#Simplefied example for embeddings
import numpy as np
import matplotlib.pyplot as plt
word_embeddings={
    "cat":[0.8,0.6],
    "kitten":[0.75,0.65],
    "dog":[0.7,0.3],
    "puppy":[0.65,0.35],
    "car":[-0.5,0.2],
    "truck":[-0.45,0.15]
}
```

```python
def cosine(vec1,vec2):
    dot_product=np.dot(vec1,vec2)
    norm_a=np.linalg.norm(vec1)#magnitude of vector 1
    norm_b=np.linalg.norm(vec2)#magnitude of vector 2

    return (dot_product/(norm_a,norm_b))
```

```python

#If close to 1 similar meanings
#if its 0 un related meanings
#if its -1 opposite meanings

cat_vector=[0.8,0.6,0.3]
kitten_vector=[0.75,0.65,0.35]
car_vector=[-0.5,0.2,0.1]

print(cosine(cat_vector,kitten_vector))
print(cosine(cat_vector,car_vector))
print(cosine(kitten_vector,car_vector))
```
