```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader=TextLoader("speech.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings=OllamaEmbeddings(model="gemma:2b")
embeddings
```

```python
#It will directly convert to embeddings and store in FAISS DB

db=FAISS.from_documents(docs,embeddings)
```

```python
#Querying

query="It is a distressing and oppressive duty"

docs=db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

```python
#As a Retriever

#We can also convert the vectorstore into a Retriever class.
#This allows us to easily use it in other langchain methods,which largely work with retriever

#A retriever is a wrapper interface provided by langchain to retrieve documents
#It internally uses similarity search but provides standard interface


retriever=db.as_retriever()#converting vector store db to retriever

docs=retriever.invoke(query)
docs
```

```python
docs[0].page_content
```

```python
#Documents with score
docs_and_score=db.similarity_search_with_score(query)

docs_and_score
```

```python
embedding_vector=embeddings.embed_query(query)

embedding_vector
```

```python
#Search with vectors


docs_score=db.similarity_search_by_vector(embedding_vector)

docs_score
```

```python
#Saving the FAISS file

db.save_local("fiass.index")
```

```python
#Loading FAISS file
# set `allow_dangerous_deserialization` to `True` to enable deserialization.
#langchain will expect same embedding model to convert new queries into vectors
new_db=FAISS.load_local("fiass.index",embeddings,allow_dangerous_deserialization=True)
docs=new_db.similarity_search(query)
```
