```python
#PDF Loader

from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("attention.pdf")
docs2=loader.load()
print(docs2)
```

```python
#RecursiveCharacterTextSplitter

#Text splitters break large docs into smaller chunks that will be retrievable individually 
# and fit within model context window limit.
#It is used for generic text.It splits based on this this list["\n\n","\n"," ",""]

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
#Chunk means document each document has 500 characters
#Some of the words may be missed if we are splitting so thats the reason we are using chunk_overlap
final_documents=text_splitter.split_documents(docs2)
```

```python
print(final_documents)
```

```python
with open("speech.txt") as f:
    speech=f.read()

print(speech)

#We convert this text to documents

text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)

text=text_splitter.create_documents([speech])#We will pass it as list


print(text)
print(text[0])
print(text[1])
```

```python
#Character Text Splitter
#This splits based on a given character sequence which defaults to "\n\n"

from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")

text=loader.load()

text
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)

#Sometimes if we \n\n is not there with in 100 characters it will run show the messages
text2=text_splitter.split_documents(text)

print(text2)
```

```python
#HTML Header Text Splitter
#HTML Header Text Splitter is structure-aware chunker that splits text at the HTML element level 
# and adds metadata for each header "relevant" to given chunk


from langchain_text_splitters import HTMLHeaderTextSplitter

html_string="""
<html>

   <body>
      <h1>Hello1</h1>
      <h2>Hello2</h2>
      <h3>Hello3</h3>
   </body>

  
</html>


"""
headers_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

text_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_splits=text_splitter.split_text(html_string)


html_splits



```

```python
url="https://docs.langchain.com/oss/python/integrations/splitters"

headers_to_split_on=[
    ("h1","Header 1")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)

html_header_splits=html_splitter.split_text_from_url(url)

html_header_splits
```

```python
#Splitting JSON Data

import json
import requests

json_data=requests.get("https://api.github.com/").json()
```

```python
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=300)

text=json_splitter.split_json(json_data)
```
